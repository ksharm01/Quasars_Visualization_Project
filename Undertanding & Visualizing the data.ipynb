```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import kde
from mpl_toolkits.mplot3d import Axes3D
import warnings
warnings.filterwarnings("ignore")
```

```python
data = pd.read_excel("datafinal2.xlsx")
size = data.shape[0]
print(size)
```

```python
x = data.RA
plt.figure(figsize=(14,7))
n, bins, patches = plt.hist(x, bins=90, facecolor='#2ab0ff', edgecolor='#e0e0e0', linewidth=0.5, alpha=0.7)
n = n.astype('int') 

for i in range(len(patches)):
    patches[i].set_facecolor(plt.cm.viridis(n[i]/max(n)))
   
plt.title('Right ascension- HISTOGRAM', fontsize=15)
plt.xlabel('RA', fontsize=12)
plt.ylabel('Values', fontsize=12)
plt.savefig("Right ascension- HISTOGRAM")
plt.show()
```

```python
x = data.DEC
plt.figure(figsize=(14,7))
n, bins, patches = plt.hist(x, bins=90, facecolor='#2ab0ff', edgecolor='#e0e0e0', linewidth=0.5, alpha=0.7)
n = n.astype('int') 

for i in range(len(patches)):
    patches[i].set_facecolor(plt.cm.viridis(n[i]/max(n)))
   
plt.title('Declination - HISTOGRAM', fontsize=15)
plt.xlabel('RA', fontsize=12)
plt.ylabel('Values', fontsize=12)
plt.savefig("Declination - HISTOGRAM")
plt.show()
```

```python
x = data.REDSHIFT
plt.figure(figsize=(14,7))
n, bins, patches = plt.hist(x, bins=90, facecolor='#2ab0ff', edgecolor='#e0e0e0', linewidth=0.5, alpha=0.7)
n = n.astype('int') 

for i in range(len(patches)):
    patches[i].set_facecolor(plt.cm.viridis(n[i]/max(n)))
   
plt.title('Redshift - HISTOGRAM', fontsize=15)
plt.xlabel('REDSHIFT', fontsize=12)
plt.ylabel('Values', fontsize=12)
plt.savefig("Redshift - HISTOGRAM")
plt.show()
```

```python
nbins=300
k = kde.gaussian_kde([data.RA,data.DEC])
xi, yi = np.mgrid[ data.RA.min():data.DEC.max():nbins*1j, data.DEC.min():data.DEC.max():nbins*1j]
zi = k(np.vstack([xi.flatten(), yi.flatten()]))

data1=pd.DataFrame({'x': xi.flatten(), 'y': yi.flatten(), 'z': zi })
fig = plt.figure(figsize=(14,7))
ax = fig.gca(projection='3d')
ax.plot_trisurf(data1.x, data1.y, data1.z, cmap=plt.cm.Spectral, linewidth=1)
ax.view_init(30, 80)

plt.savefig("Declination AND Right ascension- 3D")
```
