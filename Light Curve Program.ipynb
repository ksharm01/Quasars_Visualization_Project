```python
import pandas as pd
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings("ignore")
```

```python
# Load data from CSV file
data = pd.read_csv("2119-53792-0479_graph.csv")
```

```python
# Extract data
mjd = data['MJD']
mag = data['Mag']
magerr = data['Magerr']
```

```python
# Plotting with error bars
plt.errorbar(mjd, mag, yerr=magerr, xerr=None, fmt='.', linestyle='')

# Customize plot
plt.grid(True)  # Show grid lines
plt.gca().invert_yaxis()  # Invert y-axis
plt.gca().set_facecolor("pink")  # Set background color for the current axes
plt.xlabel("Date (MJD 55386 + days)")
plt.ylabel("V-Magnitude")
plt.title("Light Curves")
```
